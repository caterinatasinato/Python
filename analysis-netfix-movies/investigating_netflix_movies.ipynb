```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

1) **What was the most frequent movie duration in the 1990s?**

```python
# Selecting movies from dataset
movies_df = netflix_df[netflix_df["type"] == "Movie"]
movies_90s_df = movies_df[(movies_df["release_year"] >= 1990) & (movies_df["release_year"] <= 1999)]
```

```python
# Create an histogram with the column 'duration'
plt.hist(movies_90s_df['duration'], bins=48, edgecolor='black')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Durations in the 1990s')
plt.show()
```

```python
# find the most frequent movie
most_frequent_duration = movies_90s_df['duration'].mode().values[0]

# Save the anser as the integer called duration
duration = int(most_frequent_duration)

print(f"The most frequent movie duration in the 1990s is {duration} minutes.")
```

2) **Count the number of short action movies released in the 1990s**

```python
# Filter for genre 'Action'
action_movies_90s_df = movies_90s_df[movies_90s_df['genre'].str.contains('Action', na=False)]

# Count the number of short action movies released in the 1990s
short_movie_count = 0
for index, row in action_movies_90s_df.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1

print(f"The number of short action movies released in the 1990s is {short_movie_count}.")
```
